```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import glob
import geopandas as gpd
from p_tqdm import p_map
import cartopy.crs as ccrs
import cartopy
import cartopy.feature as cfeature
from matplotlib import gridspec
```

# Load data

```python
# load cci lakes
cci_lake_list = pd.read_csv("../data/cci_lakes_hydrolake_depth.csv")["CCI ID"].to_numpy()
lakes_with_clz = pd.read_csv("../data/lakes_with_climate_zone.csv", index_col = 0).loc[cci_lake_list]
cci_lakes = pd.read_csv("../data/ESA_CCI_static_lake_mask_v2_1km_UoR_metadata_fv2.1_06Oct2021_4laketemp.csv", index_col=0).loc[cci_lake_list]
cci_lakes = pd.concat([cci_lakes, lakes_with_clz.loc[:,["climate_zone"]]], axis = 1)
cci_lakes_gdf = gpd.GeoDataFrame(cci_lakes, geometry=gpd.points_from_xy(cci_lakes['LON CENTRE'], cci_lakes['LAT CENTRE']),
                                crs="epsg:4326")
cci_lakes_gdf.index.name = "cci_lake_id"
```

```python
train_stats = pd.read_csv("../data/sim_stats_trainperiod_ensemble_mean.csv", index_col = 0)
val_stats = pd.read_csv("../data/sim_stats_valperiod_ensemble_mean.csv", index_col = 0)
total_stats = pd.read_csv("../data/sim_stats_totalperiod_ensemble_mean.csv", index_col = 0)
train_stats_obs = pd.read_csv("../data/obs_stats_train.csv", index_col = 0)
val_stats_obs = pd.read_csv("../data/obs_stats_val.csv", index_col = 0)
total_stats_obs = pd.read_csv("../data/obs_stats_total.csv", index_col = 0)
```

```python
# Total period, only for observations
stat_diff_total = pd.DataFrame([], index = total_stats_obs.index)
stat_diff_total["obs_diff_mean"] = total_stats_obs["obs_cloud_mean"] - total_stats_obs["obs_mean"]
stat_diff_total["obs_diff_warmest_mean"] = total_stats_obs["obs_cloud_warmest_mean"] - total_stats_obs["obs_warmest_mean"]
stat_diff_total["obs_diff_coldest_mean"] = total_stats_obs["obs_cloud_coldest_mean"] - total_stats_obs["obs_coldest_mean"]

# Total period air2water simulations
stat_diff_total["a2w_diff_mean"] = total_stats["a2w_cloud_mean"] - total_stats["a2w_mean"]
stat_diff_total["a2w_diff_warmest_mean"] = total_stats["a2w_cloud_warmest_mean"] - total_stats["a2w_warmest_mean"]
stat_diff_total["a2w_diff_coldest_mean"] = total_stats["a2w_cloud_coldest_mean"] - total_stats["a2w_coldest_mean"]
stat_diff_total["a2w_diff_lhw_n_events"] = total_stats["a2w_cloud_lhw_n_events"] - total_stats["a2w_lhw_n_events"]
stat_diff_total["a2w_diff_lhw_intensity"] = total_stats["a2w_cloud_lhw_intensity"] - total_stats["a2w_lhw_intensity"]
stat_diff_total["a2w_diff_lhw_duration"] = total_stats["a2w_cloud_lhw_duration"] - total_stats["a2w_lhw_duration"]
stat_diff_total["a2w_diff_ice_days_0"] = total_stats["a2w_ice_days_0"] - total_stats["a2w_cloud_ice_days_0"]
stat_diff_total["a2w_diff_ice_days_076"] = total_stats["a2w_ice_days_076"] - total_stats["a2w_cloud_ice_days_076"]

# Total period, only for lstmervations
stat_diff_total["lstm_diff_mean"] = total_stats["lstm_cloud_mean"] - total_stats["lstm_mean"]
stat_diff_total["lstm_diff_warmest_mean"] = total_stats["lstm_cloud_warmest_mean"] - total_stats["lstm_warmest_mean"]
stat_diff_total["lstm_diff_coldest_mean"] = total_stats["lstm_cloud_coldest_mean"] - total_stats["lstm_coldest_mean"]
stat_diff_total["lstm_diff_lhw_n_events"] = total_stats["lstm_cloud_lhw_n_events"] - total_stats["lstm_lhw_n_events"]
stat_diff_total["lstm_diff_lhw_intensity"] = total_stats["lstm_cloud_lhw_intensity"] - total_stats["lstm_lhw_intensity"]
stat_diff_total["lstm_diff_lhw_duration"] = total_stats["lstm_cloud_lhw_duration"] - total_stats["lstm_lhw_duration"]
stat_diff_total["lstm_diff_ice_days_0"] = total_stats["lstm_ice_days_0"] - total_stats["lstm_cloud_ice_days_0"]
stat_diff_total["lstm_diff_ice_days_076"] = total_stats["lstm_ice_days_076"] - total_stats["lstm_cloud_ice_days_076"]
```

```python
# train period, only for observations
stat_diff_train = pd.DataFrame([], index = train_stats_obs.index)
stat_diff_train["obs_diff_mean"] = train_stats_obs["obs_cloud_mean"] - train_stats_obs["obs_mean"]
stat_diff_train["obs_diff_warmest_mean"] = train_stats_obs["obs_cloud_warmest_mean"] - train_stats_obs["obs_warmest_mean"]
stat_diff_train["obs_diff_coldest_mean"] = train_stats_obs["obs_cloud_coldest_mean"] - train_stats_obs["obs_coldest_mean"]

# train period air2water simulations
stat_diff_train["a2w_diff_mean"] = train_stats["a2w_cloud_mean"] - train_stats["a2w_mean"]
stat_diff_train["a2w_diff_warmest_mean"] = train_stats["a2w_cloud_warmest_mean"] - train_stats["a2w_warmest_mean"]
stat_diff_train["a2w_diff_coldest_mean"] = train_stats["a2w_cloud_coldest_mean"] - train_stats["a2w_coldest_mean"]

# train period, only for lstmervations
stat_diff_train["lstm_diff_mean"] = train_stats["lstm_cloud_mean"] - train_stats["lstm_mean"]
stat_diff_train["lstm_diff_warmest_mean"] = train_stats["lstm_cloud_warmest_mean"] - train_stats["lstm_warmest_mean"]
stat_diff_train["lstm_diff_coldest_mean"] = train_stats["lstm_cloud_coldest_mean"] - train_stats["lstm_coldest_mean"]
```

```python
# val period, only for observations
stat_diff_val = pd.DataFrame([], index = val_stats_obs.index)
stat_diff_val["obs_diff_mean"] = val_stats_obs["obs_cloud_mean"] - val_stats_obs["obs_mean"]
stat_diff_val["obs_diff_warmest_mean"] = val_stats_obs["obs_cloud_warmest_mean"] - val_stats_obs["obs_warmest_mean"]
stat_diff_val["obs_diff_coldest_mean"] = val_stats_obs["obs_cloud_coldest_mean"] - val_stats_obs["obs_coldest_mean"]

# val period air2water simulations
stat_diff_val["a2w_diff_mean"] = val_stats["a2w_cloud_mean"] - val_stats["a2w_mean"]
stat_diff_val["a2w_diff_warmest_mean"] = val_stats["a2w_cloud_warmest_mean"] - val_stats["a2w_warmest_mean"]
stat_diff_val["a2w_diff_coldest_mean"] = val_stats["a2w_cloud_coldest_mean"] - val_stats["a2w_coldest_mean"]

# val period, only for lstmervations
stat_diff_val["lstm_diff_mean"] = val_stats["lstm_cloud_mean"] - val_stats["lstm_mean"]
stat_diff_val["lstm_diff_warmest_mean"] = val_stats["lstm_cloud_warmest_mean"] - val_stats["lstm_warmest_mean"]
stat_diff_val["lstm_diff_coldest_mean"] = val_stats["lstm_cloud_coldest_mean"] - val_stats["lstm_coldest_mean"]
```

```python
stats_gdf_total = gpd.GeoDataFrame(pd.concat([cci_lakes_gdf, stat_diff_total], axis = 1))
stats_gdf_val = gpd.GeoDataFrame(pd.concat([cci_lakes_gdf, stat_diff_val], axis = 1))
stats_gdf_train = gpd.GeoDataFrame(pd.concat([cci_lakes_gdf, stat_diff_train], axis = 1))
```

# 1. Tmean

```python
fig = plt.figure(figsize = (10,6))
nt_proj = ccrs.PlateCarree()

# Reproject stats
stats_gdf_val = stats_gdf_val.to_crs(nt_proj)
stats_gdf_train = stats_gdf_train.to_crs(nt_proj)
stats_gdf_total = stats_gdf_total.to_crs(nt_proj)

# Define gridspec layout
gs = gridspec.GridSpec(2, 2, width_ratios=[1.0, 0.4], height_ratios=[1, 1])

# a2w, map
ax1 = fig.add_subplot(gs[0, 0], projection=nt_proj)
ax1.add_feature(cartopy.feature.OCEAN, zorder=0)
ax1.add_feature(cartopy.feature.LAND, zorder=0, edgecolor='black')
ax1.set_global()
gl1 = ax1.gridlines(crs=nt_proj, draw_labels=True)
gl1.top_labels = False
gl1.right_labels = False

stats_gdf_total.dropna(subset="a2w_diff_mean").plot(
    column="a2w_diff_mean",
    markersize=10,
    cmap="bwr",
    vmin=-0.20,
    vmax=0.20,
    legend_kwds={'label': '$Bias (^\\circ C)$', 
                 "orientation": "vertical", 
                 "extend": "both",
                 "shrink": .85},
    legend=True,
    ax=ax1)

# a2w, histogram
ax2 = fig.add_subplot(gs[0, 1])
sns.histplot(x="a2w_diff_mean", data=stats_gdf_total, ax=ax2, stat="proportion", bins = 20)
ax2.set_ylabel("Proportion")
ax2.set_xlabel("$Bias (^{\\circ}C)$")

# lstm, map
ax3 = fig.add_subplot(gs[1, 0], projection=nt_proj)
ax3.add_feature(cartopy.feature.OCEAN, zorder=0)
ax3.add_feature(cartopy.feature.LAND, zorder=0, edgecolor='black')
ax3.set_global()
gl3 = ax3.gridlines(crs=nt_proj, draw_labels=True)
gl3.top_labels = False
gl3.right_labels = False

stats_gdf_total.dropna(subset="lstm_diff_mean").plot(
    column="lstm_diff_mean",
    markersize=10,
    cmap="bwr",
    vmin=-0.20,
    vmax=0.20,
    legend_kwds={'label': '$Bias (^\\circ C)$', 
                 "orientation": "vertical", 
                 "extend": "both",
                 "shrink": .85},
    legend=True,
    ax=ax3)

# lstm, histogram
ax4 = fig.add_subplot(gs[1, 1])
sns.histplot(x="lstm_diff_mean", data=stats_gdf_total, ax=ax4, stat="proportion", bins = 20)
ax4.set_ylabel("Proportion")
ax4.set_xlabel("$Bias (^{\\circ}C)$")

# share x
ax4.set_xlim(-0.65, 0.3)
ax2.sharex(ax4)


# set column titles
fig.text(0.005, 0.75, "Air2water", ha='left', va='center', fontsize=11, fontweight='bold')
fig.text(0.005, 0.28, "LSTM", ha='left', va='center', fontsize=11, fontweight='bold')
fig.text(0.25, 0.95, "Map of Tmean bias", ha='left', va='center', fontsize=11, fontweight='bold')
fig.text(0.75, 0.95, "Histogram of Tmean bias", ha='left', va='center', fontsize=11, fontweight='bold')
ax1.set_title("(a1)", y=1.0, x= 0.05, pad = -10, fontsize=11, )
ax2.set_title("(a2)", y=1.0, x= 0.1, pad = -10, fontsize=11, )
ax3.set_title("(b1)", y=1.0, x= 0.05, pad = -10, fontsize=11, )
ax4.set_title("(b2)", y=1.0, x= 0.1, pad = -10, fontsize=11, )

ax1.set_ylabel("Air2water")
plt.tight_layout(rect=[0.05, 0, 1, 0.95])  # leave space for top row title
```

## 1. Tmean, Twarm and Tcold

```python
import matplotlib as mpl

fig = plt.figure(figsize=(13, 9))
nt_proj = ccrs.PlateCarree()

# Reproject once
stats_gdf_val = stats_gdf_val.to_crs(nt_proj)
stats_gdf_train = stats_gdf_train.to_crs(nt_proj)
stats_gdf_total = stats_gdf_total.to_crs(nt_proj)

# Define layout
gs = gridspec.GridSpec(3, 3, width_ratios=[1, 1, 0.03], height_ratios=[1, 1, 1], wspace=0.05)

def add_map(ax, column, cmap, vmin, vmax, title):
    ax.add_feature(cartopy.feature.OCEAN, zorder=0)
    ax.add_feature(cartopy.feature.LAND, zorder=0, edgecolor='black')
    ax.set_global()
    gl = ax.gridlines(crs=nt_proj, draw_labels=True)
    gl.top_labels = False
    gl.right_labels = False

    stats_gdf_total.dropna(subset=[column]).plot(
        column=column,
        markersize=10,
        cmap=cmap,
        vmin=vmin,
        vmax=vmax,
        legend=False,
        ax=ax,
    )
    ax.set_title(title, fontsize=12, loc='center')

# Row 1: Tmean
ax1 = fig.add_subplot(gs[0, 0], projection=nt_proj)
ax2 = fig.add_subplot(gs[0, 1], projection=nt_proj)
add_map(ax1, "a2w_diff_mean", "bwr", -0.2, 0.2, "(a1) Air2water Tmean")
add_map(ax2, "lstm_diff_mean", "bwr", -0.2, 0.2, "(a2) LSTM Tmean")
cax1 = fig.add_subplot(gs[0, 2])
norm1 = mpl.colors.Normalize(vmin=-0.2, vmax=0.2)
cb1 = mpl.colorbar.ColorbarBase(
    cax1, cmap="bwr", norm=norm1, orientation="vertical", extend='both'
)
cb1.set_label('Bias of Tmean $(^\\circ C)$')

# Row 2: Twarm
ax3 = fig.add_subplot(gs[1, 0], projection=nt_proj)
ax4 = fig.add_subplot(gs[1, 1], projection=nt_proj)
add_map(ax3, "a2w_diff_warmest_mean", "rainbow", -0.51, 0.51, "(b1) Air2water Twarm")
add_map(ax4, "lstm_diff_warmest_mean", "rainbow", -0.51, 0.51, "(b2) LSTM Twarm")
cax2 = fig.add_subplot(gs[1, 2])
norm2 = mpl.colors.Normalize(vmin=-0.51, vmax=0.51)
cb2 = mpl.colorbar.ColorbarBase(
    cax2, cmap="rainbow", norm=norm2, orientation="vertical", extend='both'
)
cb2.set_label('Bias of Twarm $(^\\circ C)$')

# Row 3: Tcold
ax5 = fig.add_subplot(gs[2, 0], projection=nt_proj)
ax6 = fig.add_subplot(gs[2, 1], projection=nt_proj)
add_map(ax5, "a2w_diff_coldest_mean", "coolwarm", -0.1, 0.1, "(c1) Air2water Tcold")
add_map(ax6, "lstm_diff_coldest_mean", "coolwarm", -0.1, 0.1, "(c2) LSTM Tcold")
cax3 = fig.add_subplot(gs[2, 2])
norm3 = mpl.colors.Normalize(vmin=-0.1, vmax=0.1)
cb3 = mpl.colorbar.ColorbarBase(
    cax3, cmap="coolwarm", norm=norm3, orientation="vertical", extend='both'
)
cb3.set_label('Bias of Tcold $(^\\circ C)$')

plt.tight_layout()
```

## 2. Heat wave metrics: N and I

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import cartopy.crs as ccrs
import cartopy.feature
from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib as mpl

# Set projection
nt_proj = ccrs.PlateCarree()

# Reproject stats
stats_gdf_val = stats_gdf_val.to_crs(nt_proj)
stats_gdf_train = stats_gdf_train.to_crs(nt_proj)
stats_gdf_total = stats_gdf_total.to_crs(nt_proj)

# Create 3x2 subplots
fig, axs = plt.subplots(3, 2, figsize=(12, 9), subplot_kw={'projection': nt_proj})

# Helper function to apply consistent map styling
def setup_map(ax):
    ax.add_feature(cartopy.feature.OCEAN, zorder=0)
    ax.add_feature(cartopy.feature.LAND, zorder=0, edgecolor='black')
    ax.set_global()
    gl = ax.gridlines(crs=nt_proj, draw_labels=True)
    gl.top_labels = False
    gl.right_labels = False

# --- Row 1: N(LHW) bias ---
for ax in axs[0]:
    setup_map(ax)

stats_gdf_total.dropna(subset="a2w_diff_lhw_n_events").plot(
    column="a2w_diff_lhw_n_events", markersize=10, cmap="viridis", vmin=-5.1, vmax=5.1,
    ax=axs[0, 0], legend=False
)
axs[0, 0].set_title("(a) Air2water: N(LHW) bias")

stats_gdf_total.dropna(subset="lstm_diff_lhw_n_events").plot(
    column="lstm_diff_lhw_n_events", markersize=10, cmap="viridis", vmin=-5.1, vmax=5.1,
    ax=axs[0, 1], legend=False
)
axs[0, 1].set_title("(b) LSTM: N(LHW) bias")

# Colorbar for row 1
cbar_ax_n = fig.add_axes([0.92, 0.69, 0.015, 0.22])
sm_n = mpl.cm.ScalarMappable(cmap="viridis", norm=mpl.colors.Normalize(vmin=-5.1, vmax=5.1))
sm_n._A = []
fig.colorbar(sm_n, cax=cbar_ax_n, label='N(LHW) Bias', extend = "both")

# --- Row 2: I(LHW) bias ---
for ax in axs[1]:
    setup_map(ax)

stats_gdf_total.dropna(subset="a2w_diff_lhw_intensity").plot(
    column="a2w_diff_lhw_intensity", markersize=10, cmap="coolwarm", vmin=-0.5, vmax=0.5,
    ax=axs[1, 0], legend=False
)
axs[1, 0].set_title("(c) Air2water: I(LHW) bias")

stats_gdf_total.dropna(subset="lstm_diff_lhw_intensity").plot(
    column="lstm_diff_lhw_intensity", markersize=10, cmap="coolwarm", vmin=-0.5, vmax=0.5,
    ax=axs[1, 1], legend=False
)
axs[1, 1].set_title("(d) LSTM: I(LHW) bias")

# Colorbar for row 2
cbar_ax_i = fig.add_axes([0.92, 0.39, 0.015, 0.22])
sm_i = mpl.cm.ScalarMappable(cmap="coolwarm", norm=mpl.colors.Normalize(vmin=-0.5, vmax=0.5))
sm_i._A = []
fig.colorbar(sm_i, cax=cbar_ax_i, label='I(LHW) Bias (°C)',extend = "both")

# --- Row 3: D(LHW) bias ---
for ax in axs[2]:
    setup_map(ax)

stats_gdf_total.dropna(subset="a2w_diff_lhw_duration").plot(
    column="a2w_diff_lhw_duration", markersize=10, cmap="Spectral", vmin=-1, vmax=1,
    ax=axs[2, 0], legend=False
)
axs[2, 0].set_title("(e) Air2water: D(LHW) bias")

stats_gdf_total.dropna(subset="lstm_diff_lhw_duration").plot(
    column="lstm_diff_lhw_duration", markersize=10, cmap="Spectral", vmin=-1, vmax=1,
    ax=axs[2, 1], legend=False
)
axs[2, 1].set_title("(f) LSTM: D(LHW) bias")

# Colorbar for row 3
cbar_ax_d = fig.add_axes([0.92, 0.09, 0.015, 0.22])
sm_d = mpl.cm.ScalarMappable(cmap="Spectral", norm=mpl.colors.Normalize(vmin=-1, vmax=1))
sm_d._A = []
fig.colorbar(sm_d, cax=cbar_ax_d, label='D(LHW) Bias (days)', extend = "both")

plt.tight_layout(rect=[0, 0, 0.9, 1])
plt.show()
```
