```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
import geopandas as gpd
```

```python
# load cci lakes
cci_lake_list = pd.read_csv("../data/cci_lakes_hydrolake_depth.csv")["CCI ID"].to_numpy()
cci_lakes = pd.read_csv("../data/ESA_CCI_static_lake_mask_v2_1km_UoR_metadata_fv2.1_06Oct2021_4laketemp.csv", index_col=0).loc[cci_lake_list]
cci_lakes_gdf = gpd.GeoDataFrame(cci_lakes, geometry=gpd.points_from_xy(cci_lakes['LON CENTRE'], cci_lakes['LAT CENTRE']),
                                crs="epsg:4326")
cci_lakes_gdf.index.name = "cci_lake_id"

# change here to replace with cloud gapped lake surface water temperature for calibration
df_tw = pd.read_csv("/nas/cee-hydro/laketemp_bias/era5land/water_temp.csv", index_col=0, 
                    parse_dates=True).loc[:, cci_lake_list.astype(str)]
df_tw_cloud = pd.read_csv("/nas/cee-hydro/laketemp_bias/era5land/water_temp_cloud.csv", index_col=0, 
                    parse_dates=True).loc[:, cci_lake_list.astype(str)]

train_period = pd.date_range("2003-01-01", "2017-12-31") # 15 years
val_period = pd.date_range("2018-01-01", "2023-12-31") # 6 years
total_period = pd.date_range("2003-01-01", "2023-12-31")
```

```python
df_train = df_tw.loc[train_period]
df_train_cloud = df_tw_cloud.loc[train_period]
```

```python
# cloud frequency
cloud_freq_mean = 1 - (df_train_cloud.groupby([df_train_cloud.index.year, df_train_cloud.index.month]).count()/df_train.groupby([df_train.index.year, df_train.index.month]).count()).reset_index().groupby("level_1").mean()
cloud_freq_90 = 1 - (df_train_cloud.groupby([df_train_cloud.index.year, df_train_cloud.index.month]).count()/df_train.groupby([df_train.index.year, df_train.index.month]).count()).reset_index().groupby("level_1").quantile(q = 0.90)
cloud_freq_10 = 1 - (df_train_cloud.groupby([df_train_cloud.index.year, df_train_cloud.index.month]).count()/df_train.groupby([df_train.index.year, df_train.index.month]).count()).reset_index().groupby("level_1").quantile(q = 0.10)

# monthly temperature
monthly_temp_mean = df_train.groupby([df_train.index.year,df_train.index.month ]).mean().reset_index().groupby("level_1").mean()
monthly_temp_90 = df_train.groupby([df_train.index.year,df_train.index.month ]).mean().reset_index().groupby("level_1").quantile(q = 0.90)
monthly_temp_10 = df_train.groupby([df_train.index.year,df_train.index.month ]).mean().reset_index().groupby("level_1").quantile(q = 0.10)
```

```python
lake_id = "28"
fig, ax1 = plt.subplots(figsize = (8,5))
ax2 = ax1.twinx()
ax1.fill_between(cloud_freq_mean.index, cloud_freq_10.loc[:, lake_id], cloud_freq_90.loc[:, lake_id], alpha = 0.5, color = "blue")
ax1.plot(cloud_freq_mean.loc[:, lake_id], color = "blue", label = "monthly mean cloud frequency")
ax2.fill_between(monthly_temp_mean.index, monthly_temp_10.loc[:, lake_id], monthly_temp_90.loc[:, lake_id], alpha = 0.5, color = "red")
ax2.plot(monthly_temp_mean.loc[:, lake_id], color = "red", label = "monthly mean temperature")
ax1.set_ylabel("Cloud frequency")
ax2.set_ylabel("Monthly mean temperature")
ax1.legend(loc = "upper left")
ax2.legend(loc = "upper right")
ax1.set_ylim([0, 1])
ax1.set_title(f"{cci_lakes_gdf.loc[int(lake_id), "NAME"]}; cci_id = {lake_id}; Lon = {cci_lakes_gdf.loc[int(lake_id), "LON CENTRE"]}; Lat = {cci_lakes_gdf.loc[int(lake_id), "LAT CENTRE"]}")
ax1.set_xlabel("Month")
```
