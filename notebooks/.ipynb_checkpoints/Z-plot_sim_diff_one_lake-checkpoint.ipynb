```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import glob
import geopandas as gpd
from p_tqdm import p_map
import cartopy.crs as ccrs
import cartopy
```

```python
# data paths to simulated water temperature
a2w_full_sim_dir = "/nas/cee-hydro/laketemp_bias/simulations/a2w_full_sim"
a2w_cloud_sim_dir = "/nas/cee-hydro/laketemp_bias/simulations/a2w_cloud_sim"
lstm_full_sim_dir = "/nas/cee-hydro/laketemp_bias/simulations/lstm_full_sim"
lstm_cloud_sim_dir = "/nas/cee-hydro/laketemp_bias/simulations/lstm_cloud_sim"
```

```python
lake_id = 1789
```

```python
test_period = pd.date_range("2001-01-01", "2023-12-31")
```

```python
a2w_full_sim = pd.read_csv(f"{a2w_full_sim_dir}/{lake_id}.csv", index_col = 0, parse_dates=True).loc[test_period]
a2w_cloud_sim = pd.read_csv(f"{a2w_cloud_sim_dir}/{lake_id}.csv", index_col = 0, parse_dates=True).loc[test_period]

sns.ecdfplot(a2w_full_sim.mean(axis = 1), stat = "percent", label = "a2w, full")
sns.ecdfplot(a2w_cloud_sim.mean(axis = 1), stat = "percent", label = "a2w, cloud")
plt.legend()
```

```python
sns.lineplot(a2w_full_sim.mean(axis = 1), label = "a2w, full")
sns.lineplot(a2w_cloud_sim.mean(axis = 1), label = "a2w, cloud")
```

```python
lstm_full_sim.quantile(q = 0.99).mean()
```

```python
lstm_cloud_sim.quantile(q = 0.99).mean()
```

```python
lstm_full_sim = pd.read_csv(f"{lstm_full_sim_dir}/{lake_id}.csv", index_col = 0, parse_dates=True).loc[test_period]
lstm_cloud_sim = pd.read_csv(f"{lstm_cloud_sim_dir}/{lake_id}.csv", index_col = 0, parse_dates=True).loc[test_period]
sns.ecdfplot(lstm_full_sim.mean(axis = 1), stat = "percent", label = "lstm, full")
sns.ecdfplot(lstm_cloud_sim.mean(axis = 1), stat = "percent", label = "lstm, cloud")
plt.legend()
```

```python
sns.lineplot(lstm_full_sim.mean(axis = 1), label = "lstm, full")
sns.lineplot(lstm_cloud_sim.mean(axis = 1), label = "lstm, cloud")
```
