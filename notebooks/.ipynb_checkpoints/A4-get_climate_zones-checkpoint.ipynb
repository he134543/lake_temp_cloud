```python
# load the points of cci_lakes
lake_list = pd.read_csv("../data/cci_lakes_hydrolake_depth.csv").loc[:, "CCI ID"].to_list()
gdf_lakes = pd.read_csv("../data/ESA_CCI_static_lake_mask_v2_1km_UoR_metadata_fv2.1_06Oct2021_4laketemp.csv", index_col = 0).loc[lake_list]
gdf_lakes = gpd.GeoDataFrame(gdf_lakes, geometry=gpd.points_from_xy(gdf_lakes['LON CENTRE'], gdf_lakes['LAT CENTRE']),
                                crs="epsg:4326")
```

```python
# gdf_lakes.to_file("/nas/cee-hydro/laketemp_bias/lakes_cci_v2.1.0_shp/shapefile/studylakes_points.shp")
```

```python
def get_major_climate_zone(gridcode):
    grid_code = str(gridcode)
    if grid_code.startswith("1"):
        clz = "Tropical"
    elif grid_code.startswith("2"):
        clz = "Arid"
    elif grid_code.startswith("3"):
        clz = "Temperate"
    elif grid_code.startswith("4"):
        clz = "Continental"
    elif grid_code.startswith("5"):
        clz = "Continental"
    elif grid_code.startswith("6"):
        clz = "Polar"
    else:
        clz = "Unkown"
    return clz
```

```python
clz_gdf = gpd.read_file("/nas/cee-hydro/laketemp_bias/koppen_climate_zone_v2/c1976_2000_0/c1976_2000.shp").to_crs(gdf_lakes.crs)
clz_gdf["climate_zone"] = clz_gdf.GRIDCODE.apply(get_major_climate_zone)
```

Spatial join

```python
# join with the nearest
lakes_with_clz = gpd.sjoin_nearest(gdf_lakes, clz_gdf, how = "left")
```

# Plot cloud cover impact on the number of observations

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
# load cci lakes
cci_lake_list = pd.read_csv("../data/cci_lakes_hydrolake_depth.csv")["CCI ID"].to_numpy()
lakes_with_clz = pd.read_csv("../data/lakes_with_climate_zone.csv", index_col = 0)
cci_lakes = pd.read_csv("../data/ESA_CCI_static_lake_mask_v2_1km_UoR_metadata_fv2.1_06Oct2021_4laketemp.csv", index_col=0).loc[cci_lake_list]
cci_lakes = pd.concat([cci_lakes, lakes_with_clz.loc[:,["climate_zone"]]], axis = 1)
cci_lakes_gdf = gpd.GeoDataFrame(cci_lakes, geometry=gpd.points_from_xy(cci_lakes['LON CENTRE'], cci_lakes['LAT CENTRE']),
                                crs="epsg:4326")
cci_lakes_gdf.index.name = "cci_lake_id"
```

```python
train_stats = pd.read_csv("../data/sim_stats_trainperiod_ensemble_mean.csv", index_col = 0)
val_stats = pd.read_csv("../data/sim_stats_valperiod_ensemble_mean.csv", index_col = 0)
total_stats = pd.read_csv("../data/sim_stats_totalperiod_ensemble_mean.csv", index_col = 0)
train_stats_obs = pd.read_csv("../data/obs_stats_train.csv", index_col = 0)
val_stats_obs = pd.read_csv("../data/obs_stats_val.csv", index_col = 0)
total_stats_obs = pd.read_csv("../data/obs_stats_total.csv", index_col = 0)
```

```python
# Train period
stat_diff_train = pd.DataFrame([], index = train_stats.index)
# The difference is defined as: trained with cloud trained data - daily trained data
stat_diff_train["a2w_diff_mean"] = train_stats["a2w_cloud_mean"] - train_stats["a2w_mean"]
stat_diff_train["a2w_diff_tmax"] = train_stats["a2w_cloud_tmax"] - train_stats["a2w_tmax"]

# The difference is defined as: trained with cloud trained data - daily trained data
stat_diff_train["lstm_diff_mean"] = train_stats["lstm_cloud_mean"] - train_stats["lstm_mean"]
stat_diff_train["lstm_diff_tmax"] = train_stats["lstm_cloud_tmax"] - train_stats["lstm_tmax"]

# Observational differences cloud  data - daily data
stat_diff_train["obs_diff_mean"] = train_stats_obs["obs_cloud_mean"] - train_stats_obs["obs_mean"]
stat_diff_train["obs_diff_tmax"] = train_stats_obs["obs_cloud_tmax"] - train_stats_obs["obs_tmax"]
# load missing training data
stat_diff_train["valid_proportion"] = train_stats_obs["valid_proportion"].to_numpy()
```

```python
stats_gdf_train = gpd.GeoDataFrame(pd.concat([cci_lakes_gdf, stat_diff_train], axis = 1))
```

```python
stats_gdf_train.valid_proportion.hist()
```

```python
fig = plt.figure(figsize = (10,6))
global_proj = ccrs.PlateCarree()

# change projection
gdf_lakes = stats_gdf_train.to_crs(global_proj)

# plot location
ax1 = fig.add_subplot(1, 1, 1,
                     projection = global_proj)
ax1.add_feature(cfeature.OCEAN, zorder=0)
ax1.add_feature(cfeature.LAND, zorder=0, 
                # edgecolor='black'
                )
# ax1.set_title("Study Lakes")
ax1.set_global()
ax1.gridlines(draw_labels=True,)

# Plot 
stats_gdf_train.plot(column = "valid_proportion",
        markersize=10,
        cmap="Reds_r",
        vmin=0.20,
        vmax=0.8,
        legend_kwds={'label': 'Proportion of non-cloudy days',
                 "orientation": "vertical",
                 # "extend": "both",
                 "shrink": 0.75},
        legend=True,
        ax=ax1
                    )

plt.tight_layout()
```
